```
import os
import json
import shutil
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```
root_path = '/content/drive/MyDrive/traffic_train/PREPROCESSED'
```

```
images_path = os.path.join(root_path, 'images')
meta_path = os.path.join(root_path, 'train_meta_data_no_augmented.json')
```

```
print('Number of images: ', len(os.listdir(images_path)))
```

```
with open(meta_path, 'r') as f:
  meta_data = json.load(f)

print(meta_data)
```

```
# get removed_img_ids (width < 20 or length < 20)
removed_img_ids = [x['image_id'] for x in meta_data['annotations'] if (x['bbox'][2] < 20 or x['bbox'][3] < 20)]
print('number of removed image ids: ', len(removed_img_ids))
```

```
image_ids = [x['image_id'] for x in meta_data['annotations']]
print('number of image ids: ', len(image_ids))
```

```
print('number of unique ids: ', len(set(image_ids)))
```

```
img_ids_set = set(image_ids) - set(removed_img_ids)
print('ids remaining: ', len(img_ids_set))
```

```
# copy all images into train folder
saved_folder = os.path.join(root_path, 'train')
count = 0
for x in img_ids_set:
  filename = x + '.png'
  count += 1
  print(count)
  path = os.path.join(images_path, filename)
  saved_path = os.path.join(saved_folder, filename)
  shutil.copyfile(path, saved_path)
```

```
# create annotation file
li = []
for x in meta_data['annotations']:
  if x['image_id'] in img_ids_set:
    path = 'drive/My Drive/traffic_train/PREPROCESSED/train/' + x['image_id'] + '.png'
    Xmin = x['bbox'][0]
    Ymin = x['bbox'][1]
    Xmax = x['bbox'][0] + x['bbox'][2]
    Ymax = x['bbox'][1] + x['bbox'][3]
    category_id = x['category_id']
    li.append({'path': path, 'Xmin': Xmin, 'Ymin': Ymin, 'Xmax': Xmax, 'Ymax': Ymax, 'category_id': category_id})

print(li)
```

```
df = pd.DataFrame(li)
df.head()
```

```
annotation_path = os.path.join(root_path, 'train_annotation.txt')
df.to_csv(annotation_path, header=None, index=None, sep=',', mode='a')
```

## Check the label

```
for x in li[:15]:
  filepath = x['path']
  img = cv2.imread(filepath)
  
  cv2.rectangle(img,(x['Xmin'], x['Ymin']), (x['Xmax'], x['Ymax']), (255, 0, 0), 1)

  plt.figure(figsize=(10,10))
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
  plt.show()
```
